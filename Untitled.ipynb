```python
df = pd.read_csv('Fixed train.csv')
words = set(stopwords.words('english'))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vocab = Counter()

for sentence in df.Phrase:
    for words in sentence.split(' '):
        vocab[words] += 1

vocab.most_common(100)
```

```python
from sklearn.model_selection import train_test_split

X = df['Phrase']
y = df['Sentiment']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Preprocessing text
def preprocessor(text):
    porter = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    sentences = text.split(' ')
    data = []
    for word in sentences:
        word = re.sub("n't", 'not', word)
        word = re.sub('\d+','',word) #Removing digit
        word = re.sub(r'[^\w\s]','',word) #Removing non-alphanumerical words 
        word = word.lower()
        word = re.sub('(rrb)|(lrb)','', word)
        word = lemmatizer.lemmatize(word, pos = 'v')
        word = lemmatizer.lemmatize(word, pos = 'n')
        data.append(word)
    return ' '.join(data)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize('better', pos = 'a') #Example of lemmatizer 
```

```python
print(preprocessor("why can n't you believe in Me!?")) #Testing preprocessing 
```

```python
count = CountVectorizer(ngram_range = (1,2), preprocessor = preprocessor, max_df = 9000,
                       strip_accents = 'unicode')

clf = Pipeline([('vect', count),
               ('clf', LogisticRegression(solver = 'lbfgs', random_state = 0, max_iter = 500, class_weight = {0: 1.05, 2: 0.90, 3:1.02, 4:1.15}))])

clf.fit(X_train, y_train)
```

```python
y_hat = clf.predict(X_test)
print(accuracy_score(y_hat, y_test))
```
