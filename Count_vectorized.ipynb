```python
import pandas as pd
import re
from nltk.stem import PorterStemmer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Fixed train.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
words = stopwords.words('english')
```

```python
def numbers(text):
    words = re.sub('\d+',' ',text)
    return words

def punctuation(text):
    words = re.sub(r'[^\w\s]',' ',text)
    return words

def capitalize(text):
    return text.lower()

def singularity(text):
    word = re.sub('\s\w\s',' ', text)
    #word = re.sub('\w\s',' ', text)
    #word = re.sub('\s\w',' ', text)
    return word

def double(text):
    word = re.sub('\s\w{2}\s',' ',text)
    #word = re.sub('\w{2}\s',' ', text)
    #word = re.sub('\s\w{2}',' ', text)
    return word 

# def stop(text):
#     if text not in stopwords.words('english'):
#         return text
#     else:
#         return ''
```

```python
df['Changed'] = df['Phrase'].apply(numbers)
df['Changed'] = df['Changed'].apply(punctuation)
df['Changed'] = df['Changed'].apply(capitalize)
df['Changed'] = df['Changed'].apply(singularity)
df['Changed'] = df['Changed'].apply(double)
# df['Changed'] = df['Changed'].apply(stop_words)
```
